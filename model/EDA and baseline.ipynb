```python
import pandas as pd
import numpy as np
from simhash import Simhash,SimhashIndex
import jieba
```

```python
sub = pd.read_csv('../input/smp_sample.csv')
```

```python
train = pd.read_table('../input/origin.txt',header=None)
train.columns = ['id','content']

test = pd.read_table('../input/target.txt',header=None)
test.columns = ['id','content']

train['source']= "train"
test['source'] = "test"
data = pd.concat([train,test])
```

```python
stopwords_path = '../input/stop_words.txt' 
```

```python
def jiebaclearText(text):
    mywordlist = []
    seg_list = jieba.cut(text, cut_all=False)
    liststr="/ ".join(seg_list)
    f_stop = open(stopwords_path)
    try:
        f_stop_text = f_stop.read( )
        #f_stop_text=unicode(f_stop_text,'utf-8')
    finally:
        f_stop.close( )
    f_stop_seg_list=f_stop_text.split('\n')
    for myword in liststr.split('/'):
        if not(myword.strip() in f_stop_seg_list) and len(myword.strip())>1:
            mywordlist.append(myword)
    return ''.join(mywordlist)
```

```python
data.head()['content'].apply(lambda x:jiebaclearText(str(x)))
```

```python
data['content'] = data['content'].apply(lambda x:jiebaclearText(str(x)))
```

```python
train = data.loc[data['source']=='train']
test = data.loc[data['source']=='test']

train.drop('source',axis=1,inplace=True)
test.drop(['source',],axis=1,inplace=True)
```

```python
train.head()
```

```python
objs = [(row["id"], Simhash(row["content"])) for index, row in train.iterrows()]
```

```python
index = SimhashIndex(objs, k=16)
```

```python
test.head(1)['content'].apply(lambda x:index.get_near_dups(Simhash(x)))
```

```python
test['result'] = test['content'].apply(lambda x:index.get_near_dups(Simhash(x)))
```
